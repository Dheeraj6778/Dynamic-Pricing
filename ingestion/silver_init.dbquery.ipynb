```sql
create table if not exists pipeline_state(
  pipeline_name string,
  last_run_ts timestamp
)
using delta;

INSERT INTO pipeline_state
SELECT 'usgs_silver', TIMESTAMP '1900-01-01 00:00:00'
WHERE NOT EXISTS (
  SELECT 1 FROM pipeline_state WHERE pipeline_name='usgs_silver'
);

CREATE TABLE IF NOT EXISTS usgs_earthquakes (
        event_id STRING,
        event_ts TIMESTAMP,
        event_date DATE,
        updated_ts TIMESTAMP,
        status STRING,
        magnitude DOUBLE,
        mag_type STRING,
        felt INT,
        cdi DOUBLE,
        mmi DOUBLE,
        tsunami INT,
        latitude DOUBLE,
        longitude DOUBLE,
        place STRING,
        dmin DOUBLE,
        rms DOUBLE,
        gap INT,
        nst INT,
        sig DOUBLE
    )
    USING DELTA
    PARTITIONED BY (event_date);

alter table pipeline_state
add columns (last_processed_event_date date);

select count(*)
from usgs_earthquakes;

select *
from pipeline_state;
```
